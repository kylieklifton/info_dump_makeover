# Written Analysis
# So all the land, including every country and every border we've ever fought over is about 30% of Earth.
# Territorial waters add 27%. 
# But international waters? 43%. The biggest chunk (being the OCEAN) belongs to no one.
                                                                                                         


```python
# Coding
                                                                                                                                  
land_only = df[df['category'] == 'land'].sort_values('land_percent', ascending=False)                                                            
land_only       
```

```python
# Focusing on the land                                                                                                                      
                                                                                                                                                   
# Russia is the biggest at 3.34%, but that's still only 3% of Earth. The top 10 countries together are about 13%. All those borders for 13%
```

```python
# Coding
top_10 = land_only.head(10)
top_10['land_percent'].sum()     
```
